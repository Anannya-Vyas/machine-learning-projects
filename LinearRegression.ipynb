# Models Are Approximations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
# Create fake house size data (in square feet)
sizes = np.array([500, 800, 1000, 1200, 1500, 1800, 2000]).reshape(-1, 1)
print("sizes:\n ",sizes)

# Corresponding prices (in thousands)
prices = np.array([50, 80, 100, 115, 150, 180, 200])
print("prices:\n ",prices)
```

```python
# visualzation
plt.scatter(sizes, prices)
plt.xlabel("House Size")
plt.ylabel("Price")
plt.show()
```

```python
# trained your first machine learning model
model = LinearRegression()
model.fit(sizes, prices)
```

```python
user_input = float(input("Enter house size: "))

predicted_price = model.predict([[user_input]])

print("Predicted price:", predicted_price)
```

```python
#extrapolation
plt.scatter(sizes, prices)
plt.plot(sizes, model.predict(sizes), color="red")
plt.show()
```

Mean Squared Error (MSE)

This is what the model tries to minimize.

```python
predictions = model.predict(sizes)

errors = prices - predictions

print("Errors:", errors)
print("Mean Squared Error:", np.mean(errors**2))
```
